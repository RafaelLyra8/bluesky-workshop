```python
from bluesky import RunEngine
from bluesky.plans import scan
from bluesky.callbacks import LivePlot, LiveTable
from ophyd.sim import det, motor, motor1
```

```python
RE = RunEngine()
```

## Scan

```python
RE(
    scan([det], motor, -1, 1, num=20),
    LivePlot('det', 'motor')
)
```

```python
RE(
    scan([det], motor, -5, 5, num=20),
    LivePlot('det', 'motor')
)
```

```python
RE(
    scan([det], motor, -5, 5, num=100),
    LivePlot('det', 'motor')
)
```

```python
RE(
    scan([det], motor, -1.5, 2, motor1, 1.3, -0.75, num=20),
    LiveTable(['det', 'motor', 'motor1'])
)
```
