```python
%matplotlib qt
from bluesky import RunEngine
from bluesky.plan_stubs import mv
from bluesky.plans import scan, list_scan, grid_scan, list_grid_scan, \
    spiral, spiral_square, rel_spiral
from bluesky.callbacks.best_effort import BestEffortCallback
from ophyd.sim import det, det4, motor, motor1, motor2
from matplotlib.pyplot import ion
from bluesky.utils import install_nb_kicker
install_nb_kicker()

ion()
```

```python
RE = RunEngine()

bec = BestEffortCallback()
RE.subscribe(bec)
```

## Scan

```python
RE(
    scan([det4], motor1, -1.5, 2, motor2, 1.3, -0.75, num=20)
)
```

## List Scan

```python
RE(
    list_scan([det], motor, [-1.3, -1.2, 0, 1.8, 2], motor1, [1.3, 1.2, -0.2, -1.3, -1.4])
)
```

## Grid Scan

```python
RE(
    grid_scan([det4], motor1, -1, 1, 10, motor2, -1, 1, 10)
)
```

```python
RE(
    grid_scan([det4], motor1, -5, 5, 30, motor2, -5, 5, 30)
)
```

## List Grid Scan

```python
RE(
    list_grid_scan([det4], motor1, [-1, -0.5, 0, 0.5, 1, 1.5], motor2, [-1.5, -1, -0.5, 0, 0.5, 1])
)
```

## Spiral Scans

```python
RE(
    spiral(
        detectors=[det4],
        x_motor=motor1,
        y_motor=motor2,
        x_start=0,
        y_start=0,
        x_range=10,
        y_range=10,
        dr=5,
        dr_y=5,
        nth=36
    )
)
```

```python
RE(
    spiral(
        detectors=[det4],
        x_motor=motor1,
        y_motor=motor2,
        x_start=0,
        y_start=0,
        x_range=25,
        y_range=25,
        dr=5,
        dr_y=5,
        nth=10
    )
)
```

## Square Spiral Scan

```python
RE(
    spiral_square(
        detectors=[det4],
        x_motor=motor1,
        y_motor=motor2,
        x_center=-2,
        y_center=-2,
        x_range=10,
        y_range=10,
        x_num=20,
        y_num=20
    )
)
```

## Relative Spiral

```python
RE(mv(motor1, 3))
RE(mv(motor2, -1))
RE(
    rel_spiral(
        detectors=[det4],
        x_motor=motor1,
        y_motor=motor2,
        x_range=10,
        y_range=10,
        dr=1.5,
        dr_y=1.5,
        nth=10
    )
)
```
