```python
%matplotlib qt
from bluesky import RunEngine
from bluesky.plans import fly, adaptive_scan
from bluesky.callbacks.best_effort import BestEffortCallback
from ophyd.sim import flyer1, det, motor
from matplotlib.pyplot import ion
from bluesky.utils import install_nb_kicker
install_nb_kicker()

ion()
```

```python
RE = RunEngine()

bec = BestEffortCallback()
RE.subscribe(bec)
```

## Fly Scan

```python
RE(
    fly([flyer1])
)
```

## Adaptive Scan

```python
RE(
    adaptive_scan([det], 'det', motor, start=-1, stop=10, min_step=0.1, max_step=7, target_delta=0.1, backstep=True)
)
```

```python
RE(
    adaptive_scan([det], 'det', motor, start=-10, stop=10, min_step=0.1, max_step=20, target_delta=0.001, backstep=True)
)
```

```python
RE(
    adaptive_scan([det], 'det', motor, start=-10, stop=10, min_step=0.1, max_step=20, target_delta=0.05, backstep=True)
)
```
